```python
import subprocess
try:
    print("=== GIT BRANCH ===")
    print(subprocess.check_output("git branch", shell=True, stderr=subprocess.STDOUT, text=True))
    print("=== GIT STATUS ===")
    print(subprocess.check_output("git status", shell=True, stderr=subprocess.STDOUT, text=True))
except Exception as e:
    print("Error:", e)
```

```python
import subprocess
try:
    print("=== STAGING CHANGES ===")
    print(subprocess.check_output("git add .", shell=True, stderr=subprocess.STDOUT, text=True))
    
    print("=== COMMITTING CHANGES ===")
    commit_msg = 'fix: enforce explicit pod_id in encounter trigger to preserve RLS isolation'
    print(subprocess.check_output(f'git commit -m "{commit_msg}"', shell=True, stderr=subprocess.STDOUT, text=True))
    
    print("=== PUSHING TO GITHUB ===")
    print(subprocess.check_output("git push origin main", shell=True, stderr=subprocess.STDOUT, text=True))
except Exception as e:
    print("Error:", e)
```
